```python
from math import sin, cos, sqrt, atan2, radians
def distance(lat1,lon1,lat2,lon2):
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = ((sin(dlat/2))**2) + (cos(lat1) * cos(lat2) * ((sin(dlon/2))**2))
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    
    return distance 

distance(11.0096756,21.0122287,11.006374,21.5151681)
```
